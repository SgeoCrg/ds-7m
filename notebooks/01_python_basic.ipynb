```python
# Nombres originales (desordenados, reales)
productos_original = [
    "iPhone 15 Pro Max",
    "MacBook Air M3",
    "AirPods Pro 2nd Gen",
    "Samsung Galaxy S24 Ultra", 
    "Nike Air Force 1",
    "Adidas UltraBoost 5X",
    "PlayStation 5 Slim",
    "Logitech MX Master 3S",
    "Dyson V15 Detect",
    "Coffee Maker Pro Deluxe"
]
productos_normalizados = []

def normalizar_nombre(nombre):
    return nombre.lower().replace(" ", "_")

productos_normalizados = [normalizar_nombre(p) for p in productos_original]

def buscar(nombre):
    lista = []
    for p in productos_normalizados:
        if nombre.lower() in p:
            lista.append(p)
    if len(lista) < 1:
        lista.append(['No hay resultados'])
    return lista

print(productos_original)
print(productos_normalizados)
msg = "Producto: " + productos_original[0]
msg2 = "Producto: {}".format(productos_original[1])
msg3 = f"Producto: {productos_original[2]}"
precio = 1299.99
msg3 = msg3, f"| Precio: {precio}€"

print(msg)
print(msg2)
print(msg3)
print(buscar('Iphone'))
print(buscar('pro'))
print(buscar('windows'))
```
